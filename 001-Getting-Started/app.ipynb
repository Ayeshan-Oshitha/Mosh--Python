# Python Implementation

##### ðŸ”¹ PyPy
- Faster alternative to CPython.

- Often significantly **faster** than CPython

# How Python Code Is Executed ?

- Source Code (.py): You write Python code in a .py file.
- Compilation to Bytecode: The code is compiled into bytecode (.pyc) by the Python interpreter.
- Execution by Virtual Machine: The bytecode is run by the Python Virtual Machine (PVM).
- Optional Optimization: In some implementations (like PyPy), a JIT compiler may optimize and compile parts to machine code for better performance.
